```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url='https://www.amazon.in/s?k=iphones&crid=3ILKVRDFEA4DT&sprefix=iphones%2Caps%2C208&ref=nb_sb_noss_2'
```

```python
Headers = ({'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/124.0.0.0 Safari/537.36'})
```

```python
webpage=requests.get(url, headers=Headers)
```

```python
webpage
```

```python
type(webpage.content)
```

```python
soup=BeautifulSoup(webpage.content)
soup
```

```python
product_names=[]
amount=[]
price=[]
Rate=[]
Discount=[]
```

```python
name=soup.find_all('span',class_='a-size-medium a-color-base a-text-normal')

for i in name:
    description =i.text
    product_names.extend([description ])

print(product_names)
```

```python
len(product_names)
```

```python
Actual_price=soup.find_all('span',class_='a-price a-text-price')
```

```python
for i in Actual_price:
    
    price1=i.find('span',class_='a-offscreen')
    ABCD=price1.text
    amount.extend([ABCD])

print(amount)
```

```python
len(amount)
```

```python
Main_price=soup.find_all('span',class_='a-price-whole')
```

```python
for i in Main_price:
    Prices=i.text
    price.extend([Prices])

print(price)
    
```

```python
len(price)
```

```python
phone_rating=soup.find_all('span',class_='a-icon-alt')
```

```python
phone_rating

for i in phone_rating:
    Rating=i.text
    Rate.extend([Rating])

print(Rate)
```

```python
Proj1={'Names':product_names,'Amount':amount,'Price':price,'Phone_Rating':Rate}
```

```python
df=pd.DataFrame(Proj1)
```
